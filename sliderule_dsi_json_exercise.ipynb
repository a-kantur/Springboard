## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# Upon exploring 'mjtheme_namecode' column it turns out that each row in this column
# represent a list of nested dictionaries with 'name' being one of the keys
# (and sometimes the values for this key are empty '')

list(df['mjtheme_namecode'][0:3])
```

```python
# Answer: 
# To answer the question, we need to:
# 1) load the json file as a string
# 2) use json_normalize to create a table with columns being keys of the nested dictionaries 
# and corresponding id-s  
# 3) count the unique values of the 'name' column (automatically sorter in descending order)
# 4) use head() method to get top 10 theme names

js_df = json.load(open('data/world_bank_projects.json'))
major_themes = json_normalize(js_df, 'mjtheme_namecode', ['id'])
print('Explore major_themes table:\n')
print(major_themes.head(5))
print('\n\n The top 10 major project themes')
major_themes.name.value_counts().head(10)
```

### 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in

Solution 1

```python
# 1) Group the major_themes from 2 on 'code' and 'name' columns and reset index. The values
# are automatically grouped by values of 'code' column (as it used to be a part of an index)
grouped=major_themes.groupby(['code','name'])['id'].count().reset_index()
grouped
```

```python
# 2) Define a function that finds a blank string in 'name' column, replaces it with a string
# value in the next row of the 'name' column , and returns a dataframe grouped by 'code' 
# and 'name' again (to group code-name duplicative rows)
#The resulting table now includes Economic management in the top 10 major themes

def fill_blank_names(df):
    for i in range(len(df)):
        if len(df['name'][i])==0:
            df['name'][i]=str(df['name'][i+1])
    return df.groupby(['code','name'])['id'].sum().sort_values(ascending=False).head(10)
fill_blank_names(grouped)
```

Solution 2

```python
# 0) Reset grouped dataframe to include empty names again
grouped=major_themes.groupby(['code','name'])['id'].count().reset_index()
grouped.head(5)
```

```python
# 1) Filter the grouped dataframe to only include names that are not blank. 
df1=grouped[grouped['name']!=''].set_index('code')
df1.head(5)
```

```python
# 2) Now we need to reassign id counts to include the counts for blank names. We can do it by grouping the dataframe 
# by 'code' column only and concatenating the resulting series to the starting dataframe.
df2=grouped.groupby('code')['id'].sum()
df3=df1.join(df2, rsuffix='id_new',lsuffix='id_old').drop(columns='idid_old').sort_values(by='idid_new', ascending=False).head(10)                
df3
```
