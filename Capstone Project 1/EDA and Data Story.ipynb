# Data Story

```python
import pandas as pd
import numpy as np; np.random.seed(0)
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from decimal import Decimal
```

## Loading Data

```python
recipes_df=pd.read_csv('recipes_df.csv')
recipes_df.set_index('id', inplace=True)
recipes_df.head(3)
```

```python
reviews_df=pd.read_csv('reviews_df.csv')
reviews_df.set_index('recipe_id', inplace=True)
reviews_df.head(3)
```

```python
tags_matrix=pd.read_csv('tags_matrix.csv')
tags_matrix.set_index('id', inplace=True)
tags_matrix.head(3)
```

```python
ingredients_matrix=pd.read_csv('ingredients_matrix.csv')
ingredients_matrix.set_index('id', inplace=True)
ingredients_matrix.head(3)
```

### Distribution of Ratings

Let's review the distribution of ratings for all the recipes in the dataset.

```python
reviews_df.reset_index(inplace=True)
df=reviews_df.groupby('rating')['recipe_id'].count().to_frame()
df.reset_index(inplace=True)
df['Percent'] = 100*df.recipe_id/df.recipe_id.sum(axis=0)
x=df.rating
y=[round(df.Percent[n],2) for n in range(0,len(df))]

_ = plt.bar(x, y, color='orange')
_ = plt.title('Distribution of Reviews by Rating')
_ = plt.xlabel('Rating')
_ = plt.ylabel('% of Reviews')
xlocs, xlabs = plt.xticks()
_ = [plt.text(xlocs[i]+0.75, v+0.15, str(v)+'%') for i, v in enumerate(y)]
plt.savefig(fname='Distribution')
```

As we can see from the chart, over 80% of the reviews are 4 and 5 stars. 

## Ingredients and Recipe Tags Analysis

### Best and Worst Ingredients by Average Recipe Rating

Let's estimate the mean recipe rating for each ingredient.

```python
avg_rating=reviews_df.groupby('recipe_id')['rating'].mean()
ingredients_matrix_rating=ingredients_matrix.join(avg_rating, how='left')
```

```python
ingredient_ratings=[]
for ingredient in ingredients_matrix_rating.columns[:-1]:
    ingredient_ratings.append(ingredients_matrix_rating.rating[ingredients_matrix_rating[ingredient]==1].mean())
ingredient_ratings_series=pd.Series(index=ingredients_matrix_rating.columns[:-1],
                                    data=ingredient_ratings,
                                    name='avg_rating')
```

Let's look at the top and bottom 10 ingredients by mean recipe ratings:

```python
df=ingredient_ratings_series.to_frame()
df['ingredient'] = df.index
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df.sort_values(by='avg_rating',inplace=True, ascending=False)
df1=df.head(10)
df2=df.tail(10)
```

```python
f, axes = plt.subplots(1, 2,sharey='row', figsize=(15,7.5))

ymin = min(df2.avg_rating-0.01)
ymax = max(df1.avg_rating+0.01)

_ = sns.barplot(x='ingredient', y='avg_rating', data=df1, color='blue',orient='v', ax=axes[0]).set_title('Top 10')
axes[0].tick_params(axis='x', labelrotation = 90)
axes[0].set(xlabel='Ingredients', ylabel='Average Rating')
for p in axes[0].patches:
    axes[0].annotate(round(p.get_height(),3), (p.get_x(), p.get_height()))


_ = sns.barplot(x='ingredient', y='avg_rating', data=df2, color='blue',orient='v', ax=axes[1]).set_title('Bottom 10')
plt.ylabel('Average Rating')
plt.xlabel('Ingredients')
axes[1].tick_params(axis='x', labelrotation = 90)
axes[1].set(xlabel='Ingredients', ylabel=' ')
for p in axes[1].patches:
    axes[1].annotate(round(p.get_height(),3), (p.get_x(), p.get_height()))

plt.suptitle('Ingredients Ranked by Average Recipe Rating')

plt.ylim(ymin,ymax)
_ = plt.savefig(fname='Ingredients Rank')
plt.show()
```

The results are: <br>
 - The best ingredient is garlic. It has the average recipe ranking of 4.453<br>
 - The worst ingredient is baking powder. It has the average recipe ranking of 4.388<br>
<br><br> (A reminder that we are looking at recipes with over 10 reviews, and we know from the previous analysis that the majority of such recipes have an average rating of 3.5 or higher. Also, we are only analyzing the ingredients present in 100 or more recipes.)

### Best and Worst Recipe Tags by Average Recipe Rating

Let's estimate the mean recipe rating for each recipe tag.

```python
avg_rating=reviews_df.groupby('recipe_id')['rating'].mean()
tags_matrix_rating=tags_matrix.join(avg_rating, how='left')
```

```python
tags_ratings=[]
for tag in tags_matrix_rating.columns[:-1]:
    tags_ratings.append(tags_matrix_rating.rating[tags_matrix_rating[tag]==1].mean())
tags_ratings_series=pd.Series(index=tags_matrix_rating.columns[:-1],
                                    data=tags_ratings,
                                    name='avg_rating')
```

Let's look at the top and bottom 10 ingredients by mean recipe ratings:

```python
df3=tags_ratings_series.to_frame()
df3.reset_index(inplace=True)
df3 = df3.rename(columns = {'index':'tag'})
df3.sort_values(by='avg_rating',inplace=True, ascending=False)
df4=df3.head(10)
df5=df3.tail(10)
```

```python
f, axes = plt.subplots(1, 2,sharey='row', figsize=(15,7.5))

ymin = min(df5.avg_rating-0.01)
ymax = max(df4.avg_rating+0.01)

_ = sns.barplot(x='tag', y='avg_rating', data=df4, color='blue',orient='v', ax=axes[0]).set_title('Top 10')
axes[0].tick_params(axis='x', labelrotation = 90)
axes[0].set(xlabel='Tags', ylabel='Average Rating')
for p in axes[0].patches:
    axes[0].annotate(round(p.get_height(),3), (p.get_x(), p.get_height()))


_ = sns.barplot(x='tag', y='avg_rating', data=df5, color='blue',orient='v', ax=axes[1]).set_title('Bottom 10')
plt.ylabel('Average Rating')
plt.xlabel('Tags')
axes[1].tick_params(axis='x', labelrotation = 90)
axes[1].set(xlabel='Tags', ylabel=' ')
for p in axes[1].patches:
    axes[1].annotate(round(p.get_height(),3), (p.get_x(), p.get_height()))

plt.suptitle('Recipe Tags Ranked by Average Recipe Rating')
plt.ylim(ymin,ymax)
_ = plt.savefig(fname='Tags Rank')
plt.show()
```

The results are: <br>
 - The best tag is 'beijing'. It has the average recipe ranking of 5.0 <br>
 - The worst tag is 'pressure-canning'. It has the average recipe ranking of 2.981<br>

### Ingredients-Recipe Tags Co-occurence

#### Best Tags + Best Ingredients

Let's review how the best rated ingredients and recipe tags occur together in recipes. We will estimate the number of recipes for each co-ocurance. We need df1 (best ingredients) and df4 (best ratings) from the previous analysis.

```python
cooc = pd.DataFrame(0,
                    columns=[i for i in list(df1.ingredient.values)], 
                    index=df4.tag)

for rid in ingredients_matrix.index:
    for ing in cooc.columns:
        for tag in cooc.index:
            if ingredients_matrix[ing][rid]==1 and tags_matrix[tag][rid]==1:
                cooc.loc[tag, ing]+=1
```

```python
cooc
```

```python
import matplotlib.colors as mcolors
plt.figure(figsize=(7,7))

cmap = mcolors.LinearSegmentedColormap.from_list("n",['white','darkgreen'])

akws = {"ha": 'center',"va": 'center', "size": 15}

ax = sns.heatmap(cooc,
                 vmin=-1, vmax=4, center=0,
                 square=True,
                 annot = True,
                 annot_kws=akws,
                 linewidth=0.2,
                 linecolor='black',
                 cmap=cmap)

_ = ax.set_xticklabels(ax.get_xticklabels(),
                      rotation=90,
                      horizontalalignment='right')

b, t = plt.ylim()
b += 0.5
t -= 0.5
_ = plt.ylim(b, t)

_ = plt.title('Best Ingredients + Best Tags', size=20)
```

The highest co-occurence is for "ragu-recipe-contest" tag with garlic and mayonnaise (4 recipes)

## Other Recipe Attributes

Now we will review the correlation between recipe ratings and the following recipe attributes: <br>
 - minutes to cook the recipe
 - number of steps in a recipe
 - number of ingredients
 - nutrition values: calories, total fat, sugars, sodium, protein, saturated fat, total carbohydrates

Let's create a dataframe for analysis by combining all the relevant columns from recipes and reviews dataframes:

```python
df=recipes_df[['minutes', 'n_steps', 'n_ingredients','Calories', 'Total Fat', 'Sugars', 'Sodium', 'Protein',
       'Saturated Fat', 'Total Carbohydrate']].join(reviews_df['rating'], how='left')
```

```python
plt.figure(figsize=(10,10))

akws = {"ha": 'center',"va": 'center'}

ax = sns.heatmap(round(df.corr(), 3),
                 vmin=-1, vmax=1, center=0,
                 square=True,
                 annot = True,
                 annot_kws=akws,
                 cmap="coolwarm")

_ = ax.set_xticklabels(ax.get_xticklabels(),
                      rotation=45,
                      horizontalalignment='right')
b, t = plt.ylim()
b += 0.5
t -= 0.5
_ = plt.ylim(b, t)

_ = plt.title('Correlation Between Recipe Attributes', size=20)
_ = plt.savefig(fname='Correlation')
```

Correlation matrix shows that there is a strong correlation between calories content and total fats (saturated fats specifically), protein and carbohydrates. However, there is no strong correlation between the recipe rating and any of the mentioned recipe attributes.

We want to further investigate the distributions of recipe attributes and recipe ratings by using boxplots. Since we will be plotting a lot, let's define a function to make boxplots of the data:

```python
def makeboxplot(df, col):
    sns.boxplot(x=df.rating, y=df[col].values, data=df)
    _ = plt.title('Distribution of ' + col + ' for each recipe rating')
    _= plt.ylabel(col)
```

```python
makeboxplot(df, 'minutes')
```

Let's remove the outlier recipe that require over 500 minutes to make.

```python
makeboxplot(df.drop(df.index[df.minutes>500]), 'minutes')
```

Without the outliers, higher rated recipes on average require more time to make than the lower rated recipes.

```python
makeboxplot(df, 'n_steps')
```

The average number of steps is close to 10 for all the ratings, however 5 star rated recipes have more outliers with over 20 steps.

```python
makeboxplot(df, 'Calories')
```

Recipes with higher ratings include more calories.

```python
makeboxplot(df, 'Total Fat')
```

```python
makeboxplot(df.drop(df.index[df['Total Fat']>200]), 'Total Fat')
```

Higher rated recipes on average have more fat.

```python
makeboxplot(df, 'Sugars')
```

```python
makeboxplot(df.drop(df.index[df.Sugars>300]), 'Sugars')
```

Higher rated recipes on average have more sugar.

```python
makeboxplot(df, 'Sodium')
```

Let's remove the outlier recipes with sodium content over 200 and revisit the boxplot:

```python
makeboxplot(df.drop(df.index[df.Sodium>200]), 'Sodium')
```

4 and 5 star recipes on average have less sodium than lower rated recipes.

```python
makeboxplot(df, 'Protein')
```

Let's remove the outlier recipes with protein content over 400:

```python
makeboxplot(df.drop(df.index[df.Protein>400]), 'Protein')
```

Without the outliers we can see that recipes rated 2 stars on average have more protein content than the higher rated recipes.

```python
makeboxplot(df, 'Saturated Fat')
```

```python
makeboxplot(df.drop(df.index[df['Saturated Fat']>200]), 'Saturated Fat')
_ = plt.savefig(fname='SatFat')
```

After removing the outlier recipes with saturated fat content over 200 we can see that higher levels of saturated fat correspond with lower ratings of 1 and 3, while lower levels correspond with the 4 and 5 star recipes.

```python
makeboxplot(df, 'Total Carbohydrate')
```

```python
makeboxplot(df.drop(df.index[df['Total Carbohydrate']>50]), 'Total Carbohydrate')
```

After removing the outlier recipes with total carbohydrates content over 50 we can see that on average the higher rated recipes contain more total carbohydrates.

Now we can summarize what successful recipes have in common: <br>
 - on average they take longer to make and include 10 or more steps
 - in terms of nutritional value, on average they have more total fats, sugars and carbohydrates, but less sodium and saturated fats, and slightly less protein.
